```python
saved_Pin = 3344
balance = 0
max_Attempt = 5
daily_withdrawl_limit = 20000
attempt = 0
auth = False
ttl_withdrawl = 0
trans_history = []

def check_pin(user_pin) :
  """ Checking the user entered PIN with already saved PIN.
  """
  return user_pin == saved_Pin

while attempt < max_Attempt :
  try :
    pin = int(input("Enter your PIN :- "))
  except ValueError :
    print("Invalid PIN , Enter again.")
    continue

  if check_pin(pin) :
      auth = True
      print("Authentication successfully.")
      break
  else :
    attempt += 1
    print("Wrong PIN.")

while True :
  print("===========***===========")
  print("ATM Menu :-")
  print("1. Balance")
  print("2. Deposit")
  print("3. Withdrawl")
  print("4. Mini Statement")
  print("5. Exit")
  print("===========***===========")

  try :
    option = int(input("Choose an option :- "))
  except ValueError :
    print("Enter Valid number.")
    continue

  if option == 1 :
    print("Your Current Balance :-" , balance)
  elif option == 2 :
    try :
      amount = int(input("Enter deposit amount :-"))
    except ValueError :
      print("Enter valid amount")
      continue

    if amount <= 0 :
      print("Amount must be greater than 0.")
    else :
      balance += amount
      trans_history.append(f"Deposited Rs.{amount}")
      print("Amount added successfully.")

  elif option == 3 :
    if balance <= 0 :
      print("Balance is not sufficient")
      continue
    try :
      amount = int(input("Enter withdrawl amount :-"))
    except ValueError :
      print("Enter valid amount.")
      continue

    if amount <= 0 :
      print("Amount must be greater than 0.")
    elif amount > balance :
      print("Insufficient Balance.")
    elif ttl_withdrawl + amount > daily_withdrawl_limit :
      print("Daily limit exceeded , Come tomorrow.")
    else :
      balance -= amount
      ttl_withdrawl += amount
      trans_history.append(f"Withdwarl Rs.{amount}")
      print("Your amount here :- Rs." , amount )

  elif option == 4 :
    print("MiniStatement :- ")
    if not trans_history :
      print("No History.")
    else :
      for i in trans_history :
        print(i)
    print("Current Balance :- " , balance)

  elif option == 5 :
    print("Thank You !")
    break
  else :
    print("Invalid Option.")

```
